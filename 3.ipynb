```python
import pandas as pd
import matplotlib.pyplot as plt
```

ПЗ

```python
df = pd.read_csv('./data3/Underlying Cause of Death, 1999-2020.txt', delimiter='\t')
df = df[['Single-Year Ages', 'Deaths']]
df.dropna(how='any')
age = []
death = []
for row in df.values:
    if row[0] == '< 1 year':
      age.append('0')
      death.append(row[1])
      continue
    if row[0] == '100+ years':
      age.append('100')
      death.append(row[1])
      continue
    if row[0] == 'Not Stated':
      break
    death.append(row[1])
    age.append(str(row[0]).split(' ')[0])

fig, ax = plt.subplots(figsize=(40, 10))
ax.bar(age,death)
plt.show()
```

```python
df = pd.read_fwf('./data3/Pogoda_oGurtsovo.txt', names = ['unnkown-1', 'year', 'month', 'day', 'hour', 'temp', 'unknown-2'])

df = df[df['year'] == 1972]

df_noon = df[df['hour'] == 12]['temp']
df_night = df[df['hour'] == 0]['temp']

fig, ax = plt.subplots()

ax.scatter(df_night, df_noon)

ax.set_xlabel('night temp')
ax.set_ylabel('day temp')

plt.show()

```
