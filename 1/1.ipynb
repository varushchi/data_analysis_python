```python
def turn(matrix, angle):
  degAngle = np.deg2rad(angle)
  turnMatrix = np.array([[np.cos(degAngle), np.sin(degAngle)], [-np.sin(degAngle), np.cos(degAngle)]])
  return np.matmul(matrix, turnMatrix)

PM1_120 = turn(P1, 120)
PM2_120 = turn(P2, 120)
PM1_240 = turn(P1, 240)
PM2_240 = turn(P2, 240)

print('120')
plt.plot(PM1_120[:,0], PM1_120[:,1])
plt.plot(PM2_120[:,0], PM2_120[:,1])
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.show()

print('240')
plt.plot(PM1_240[:,0], PM1_240[:,1])
plt.plot(PM2_240[:,0], PM2_240[:,1])
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.show()

timesFive = np.array([[5,0], [0,5]])
P1_5 = np.matmul(P1, timesFive)
P2_5 = np.matmul(P2, timesFive)

print("5 Ñ€Ð°Ð· Ð±Ð¾Ð»ÑŒÑˆÐµ")
plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(P1_5[:,0], P1_5[:,1])
plt.plot(P2_5[:,0], P2_5[:,1])
plt.show()

print("Ð¿Ð¾ 60")
for i in range(6):
  plt.plot(turn(P1_5, i*60)[:,0], turn(P1_5, i*60)[:,1])
  plt.plot(turn(P2_5, i*60)[:,0], turn(P2_5, i*60)[:,1])
plt.show()

```

```python
V = np.array([[-1,1], [4,0]])

PV1 = np.matmul(P1, V)
PV2 = np.matmul(P2, V)

plt.plot(P1[:,0], P1[:,1])
plt.plot(P2[:,0], P2[:,1])
plt.plot(PV1[:,0], PV1[:,1])
plt.plot(PV2[:,0], PV2[:,1])
plt.show()

VInv = np.linalg.inv(V)

back1 = np.matmul(PV1, VInv)
back2 = np.matmul(PV2, VInv)

plt.plot(back1[:,0], back1[:,1])
plt.plot(back2[:,0], back2[:,1])
plt.plot(PV1[:,0], PV1[:,1])
plt.plot(PV2[:,0], PV2[:,1])
plt.show()
```
